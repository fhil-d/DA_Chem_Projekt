## Add Column for Producttype Grouping

```python
# Function for adding column producttype
```

```python
def label (row):
    if row['p_product'] in ["S085 HR 4040", "S400 HR","S440 HR"]:
        return 'High Rejection'
    elif row['p_product'] in ["S085 HF 4040","S400 HF" ,"S440 HF" , "B085 HF 4040" , "B400 HF" , "B440 HF"]:
        return 'High Flow'
    elif row['p_product'] in ["B400 HP" , "B440 HP"]:
        return "High Performance"
    elif row['p_product'] in ["B085 FR 4040" , "B370 FR" , "B400 FR ASD" , "B400 FR"]:
        return "Fouling Resistant"
    elif row['p_product'] in ["B085 LE 4040" , "B400 LE" , "B400 LE ASD" , "B440 LE"]:
        return "Low Energy"
    elif row['p_product'] in ["B085 ULP 4040" ,"B400 ULP ASD" ,"B440 ULP"]:
        return "Ultra Low Pressure"

# to be added before grouping columns
```

```python
df['Product_type_group'] = df.apply (lambda row : label(row), axis = 1)
```
